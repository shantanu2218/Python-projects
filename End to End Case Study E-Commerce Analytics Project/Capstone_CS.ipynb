### Understanding monthly customer retention

```python
# using pd.crosstab to get monthly count of customer
df['Year_month_purchase'] = df.order_purchase_timestamp.sort_values().dt.strftime('%Y-%m')
user_purchase = pd.crosstab(df.customer_unique_id,df.Year_month_purchase).sort_index(axis=1,ascending=True).reset_index()
user_purchase.head(2)
```

#### Steps to calculate retention rate
    # Find out how many customers you have at the end of a given period (week, month, or quarter). 
    # Subtract the number of new customers you’ve acquired over that time. 
    # Divide by the number of customers you had at the beginning of that period. 
    # Then, multiply that by one hundred.

```python
# Using for loop to calculate total user and retained user
months = user_purchase.columns[1:]
retention_array = []
for i in range(len(months)):
    retention_data = {}
    selected_month = months[i]
    prev_month = months[i-1]
    retention_data['InvoiceYearMonth'] = (selected_month)
    retention_data['TotalUserCount'] = user_purchase[selected_month].sum()
    retention_data['RetainedUserCount'] = user_purchase[(user_purchase[selected_month]>0) & (user_purchase[prev_month]>0)][selected_month].sum()
    retention_array.append(retention_data)    
```

```python
#convert the array to dataframe and calculate Retention Rate
user_purchase = pd.DataFrame(retention_array)
user_purchase['RetentionRate'] = user_purchase['RetainedUserCount']*100/user_purchase['TotalUserCount']
```

###  Performing Customers/sellers Segmentation

####     a. Divide the customers into groups based on the revenue generated

```python
cust_seg = pd.cut(x=(df.groupby(df.customer_unique_id)['payment_value'].sum()),
                     bins=5,labels=['Very_Low_Spend','Low_Spend',
                                 'Medium_Spend','High_spend','Very_High_Spend']).reset_index()

cust_seg.rename(columns={'payment_value':'customer_segment'},inplace=True)
```

```python
# Joining it with main dataframe
df = df.merge(right=cust_seg,on='customer_unique_id',how='inner')
```

```python
df.customer_segment.value_counts()*100/df.customer_unique_id.size
# Majority of customers we have are Very low spend customers
```

#### b. Divide the sellers into groups based on the revenue generated

```python
segment_seller=pd.cut(x=df.groupby('seller_id')['payment_value'].sum(),
                      bins=5,labels=['Very_Low_Rev','Low_Rev','Medium_Rev','High_Rev','Very_High_Rev']).reset_index()

segment_seller.rename(columns={'payment_value':'seller_seg'},inplace=True)
```
